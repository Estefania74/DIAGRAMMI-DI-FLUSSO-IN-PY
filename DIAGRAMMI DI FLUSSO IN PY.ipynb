## Calcola quadrati

*Il codice chiede all'utente di inserire tre numeri, calcola il quadrato di ciascuno e stampa i risultati uno alla volta.*

```python
lista = [0] * (3)
quadrati = [0] * (3)

print("ciao, sono vicino a te... calcolo i quadrati, dammi tre numeri")
for i in range(0, 2 + 1, 1):
    lista[i] = int(input())
print("calcolo i quadrati")
for i in range(0, 2 + 1, 1):
    quadrati[i] = lista[i] ** 2
    print("i quadrati dei numeri assegnati sono:")
    print(quadrati[i])
```

## Calcola area quadrati e triangoli

*Il codice chiede all'utente se vuole calcolare l'area di un quadrato o di un triangolo, esegue il calcolo in base alla scelta e mostra il risultato.*

```python
print("Ciao, vuoi calcolare l'area del quadrato o del triangolo? Digita q per il quadrato e t per il triangolo")
scelta = input()
if scelta == "q":
    print("Dammi il lato del quadrato")
    l = float(input())
    a = l ** 2
    print("L'area della figura scelta è:")
    print(a)
else:
    if scelta == "t":
        print("Dammi la base del triangolo")
        b = float(input())
        print("Dammi l'altezza del triangolo")
        h = float(input())
        a = b * h / 2
        print("L'area della figura scelta è:")
        print(a)
    else:
        print("scemo di merda, devi scrivere q per il quadrato e t per il triangolo")
```

## Gioco carta, forbice o sasso

*Il codice permette di giocare a carta, forbice, sasso contro il computer: chiede la scelta dell'utente, genera una scelta casuale per il computer, determina il vincitore e stampa il risultato.*

```python
import random
random.seed()   #Prepare random number generator

print("ciao scegli carta, forbice o sasso")
scelta = input()
while scelta != "carta" and scelta != "forbice" and scelta != "sasso" and scelta != "esci":
    print("scemo rifai la scelta")
    scelta = input()
while scelta != "esci":
    numero = int(random.random() * 3)
    if numero == 0:
        computer = "carta"
    else:
        if numero == 1:
            computer = "forbice"
        else:
            computer = "sasso"
    if scelta == "carta" and computer == "sasso" or scelta == "forbice" and computer == "carta" or scelta == "sasso" and computer == "forbice":
        print(computer)
        print("hai vinto!!")
    else:
        if scelta == "carta" and computer == "carta" or scelta == "forbice" and computer == "forbice" or scelta == "sasso" and computer == "sasso":
            print(computer)
            print("pareggio!")
        else:
            print(computer)
            print("hai perso!!")
    scelta = input()
    while scelta != "carta" and scelta != "forbice" and scelta != "sasso" and scelta != "esci":
        print("scemo rifai la scelta")
        scelta = input()
```

## Pari o dispari

*Il codice chiede tre numeri all'utente, verifica se sono pari o dispari e poi stampa i numeri inseriti insieme al risultato della verifica (pari o dispari).*

```python
lista = [0] * (3)

print("dammi tre valori della lista")
for i in range(0, 2 + 1, 1):
    lista[i] = int(input())
for i in range(0, 2 + 1, 1):
    if lista[i] % 2 == 0:
        print("è pari")
    else:
        print("è dispari")
```

## Controllo numero

*Il codice chiede un numero, verifica se è maggiore, minore o uguale a zero, poi controlla se il suo valore assoluto è pari o dispari e stampa i risultati corrispondenti.*

```python
from math import fabs
print("dammi un numero")
numero = int(input())
if numero > 0:
    print("è maggiore di zero")
else:
    if numero == 0:
        print("il numero è zero")
    else:
        print("il numero è minore di zero")
if fabs(numero) % 2 == 0:
    print("il numero è pari")
else:
    print("il numero è dispari")
```

## Fattoriale di un numero

*Il codice calcola il fattoriale di un numero fornito dall'utente, stampando il risultato parziale a ogni passo del calcolo.*

```python
risultato = 1
print("dammi il numero del quale vuoi il fattoriale")
numero = int(input())
for i in range(1, numero + 1, 1):
    risultato = i * risultato
    print(risultato)
```

## Calcolo prodotto scalare di due vettori

*Questo codice calcola il prodotto scalare di due vettori tridimensionali. Chiede all'utente di inserire i tre elementi di ciascun vettore, poi moltiplica gli elementi corrispondenti dei due vettori e somma i prodotti per ottenere il risultato finale, che viene stampato.*

```python
lista1 = [0] * (3)
lista2 = [0] * (3)

risultato = 0
print("Dammi le componenti dei due vettori")
for i in range(0, 2 + 1, 1):
    print("Inserisci l'elemento " + str(i + 1) + " del primo vettore:")
    lista1[i] = int(input())
    print("Inserisci l'elemento " + str(i + 1) + " del secondo vettore:")
    lista2[i] = int(input())
for i in range(0, 2 + 1, 1):
    risultato = risultato + lista1[i] * lista2[i]
print("Il risultato del prodotto scalare è:" + str(risultato))
```
